````python
from ipywidgets import FileUpload
from IPython.display import display

# Create upload widget
uploader = FileUpload(accept='.py,.js,.java', multiple=True)
display(uploader)
````

````python
def get_language_tag(filename):
    if filename.endswith(".py"):
        return "python"
    elif filename.endswith(".js"):
        return "js"
    elif filename.endswith(".java"):
        return "java"
    else:
        return ""
````

````python
def build_grounded_prompt(code, filename,language_type):
    grounding_docs = {
        "Style": "https://peps.python.org/pep-0008/",
        "Security": "https://owasp.org/www-project-top-ten/",
        "Performance": "https://developer.mozilla.org/en-US/docs/Web/JavaScript/Performance",
        "Best Practices": "https://google.github.io/styleguide/"
    }

    prompt = f"""
You are a senior code reviewer. Analyze the file `{filename}` below and return a list of code quality issues.

For each issue:
- Classify it as one of: Style, Security, Performance, Best Practices
- Suggest a fix
- Include a brief explanation
- Link to a relevant official documentation page for grounding

Use the following references when applicable:
- Style: {grounding_docs['Style']}
- Security: {grounding_docs['Security']}
- Performance: {grounding_docs['Performance']}
- Best Practices: {grounding_docs['Best Practices']}

Here is the code to review with the language type:
```{language_type}
{code},"""

    return prompt
````

````python
uploaded_files = {}
grounding_prompt = ''
for file in uploader.value:
    name = file['name']
    # Convert memoryview to bytes, then decode
    content = bytes(file['content']).decode('utf-8', errors='ignore')
    uploaded_files[name] = content
    language_type = get_language_tag(name)
    grounding_prompt = build_grounded_prompt(content,name,language_type)
    print(grounding_prompt)
````
